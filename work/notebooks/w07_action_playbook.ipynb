# Ranked Actions + Reason Codes

The validated ML model is used to prioritize content improvements.

| Rank | Recommended Action | Reason Code | Priority |
|------|---------------------|-------------|----------|
| 1 | Refresh outdated content | High impressions but low CTR | High |
| 2 | Improve title and meta description | Low click-through rate | High |
| 3 | Add internal links | Weak internal linking | Medium |
| 4 | Update statistics and examples | Content is outdated | Medium |
| 5 | Continue monitoring | Stable performance | Low |

These actions help content editors focus on pages that are most likely to improve overall content performance.

#2  . Intended Use and Limits

*Who uses this, for what — and where it stops being valid.*





This playbook helps content editors prioritize pages for review using the validated machine learning model.

It is intended for research and educational purposes only. The recommendations support decision making but do not replace human judgment.

The model should not be used as the only source for business decisions because predictions are based on historical data and may not reflect future user behavior.

## 3. Human review + the no-go list

*What a person must check before acting. What should never be automated.*

# Human Review

Every recommendation should be reviewed before implementation.

Editors should verify:

- Content accuracy
- SEO quality
- Business relevance
- Brand guidelines
- User intent

# No-Go List

The following actions should NOT be automated:

- Publishing content automatically
- Deleting existing content
- Medical or legal recommendations
- Sensitive business decisions
- Final editorial approval

## 4. Monitoring / retrain triggers

*What would tell you the recommendations went stale?*

# Monitoring and Retraining

The model should be monitored regularly.

Retraining should be considered when:

- Model accuracy decreases.
- CTR changes significantly.
- User behavior changes.
- New content categories are introduced.
- Data distribution changes over time.

Regular monitoring helps keep recommendations reliable.

## 5. Exports for the paper

*Write the queue (and any figures you want to reuse) to work/outputs/ — your paper builds on these files.*

```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Create ranked action queue
action_queue = pd.DataFrame({
    "Rank":[1,2,3,4,5],
    "Action":[
        "Refresh outdated content",
        "Improve title and meta description",
        "Add internal links",
        "Update statistics and examples",
        "Monitor performance"
    ],
    "Reason Code":[
        "High impressions, Low CTR",
        "Low CTR",
        "Weak internal linking",
        "Outdated content",
        "Stable performance"
    ],
    "Priority":[
        "High",
        "High",
        "Medium",
        "Medium",
        "Low"
    ]
})

# Create folders
os.makedirs("work/outputs", exist_ok=True)
os.makedirs("work/figures", exist_ok=True)

# Export CSV
action_queue.to_csv("work/outputs/action_queue.csv", index=False)

# Create figure
priority_counts = action_queue["Priority"].value_counts()

plt.figure(figsize=(5,4))
priority_counts.plot(kind="bar")
plt.title("Action Priority")
plt.xlabel("Priority")
plt.ylabel("Number of Actions")

plt.tight_layout()
plt.savefig("work/figures/action_priority_chart.png")
plt.show()

print("✅ action_queue.csv exported to work/outputs/")
print("✅ action_priority_chart.png exported to work/figures/")
```
